```python
from scipy.io import arff
import pandas as pd

data_2c, meta_2c = arff.loadarff('column_2C_weka.arff')
data_3c, meta_3c = arff.loadarff('column_3C_weka.arff')

df_2c = pd.DataFrame(data_2c)
df_3c = pd.DataFrame(data_3c)

df_2c['class'] = df_2c['class'].map({b'NO': 0, b'AB': 1})
df_3c['class'] = df_3c['class'].map({b'NO': 0, b'AB': 1})

print(df_2c.head())
```

```python
print("Original Label Value:", df_2c['class'].unique())
```

```python
class0 = df_2c[df_2c['class'] == 0].reset_index(drop=True)
class1 = df_2c[df_2c['class'] == 1].reset_index(drop=True)

train = pd.concat([class0.iloc[:70], class1.iloc[:140]])
test = pd.concat([class0.iloc[70:], class1.iloc[140:]])

X_train = train.drop('class', axis=1)
y_train = train['class']
X_test = test.drop('class', axis=1)
y_test = test['class']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(train, hue='class', palette={0: 'blue', 1: 'red'})
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='variable', y='value', hue='class', data=pd.melt(train, id_vars='class'))
plt.xticks(rotation=45)
plt.show()
```
