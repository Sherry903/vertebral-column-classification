```python
df_2c = pd.DataFrame(data_2c)

df_2c['class'] = df_2c['class'].str.decode('utf-8').map({
    'Normal': 0,
    'Abnormal': 1
})

class0 = df_2c[df_2c['class'] == 0].reset_index(drop=True)  
class1 = df_2c[df_2c['class'] == 1].reset_index(drop=True)  

train = pd.concat([class0.iloc[:70], class1.iloc[:140]], ignore_index=True)
test = pd.concat([class0.iloc[70:], class1.iloc[140:]], ignore_index=True)
```

```python
train = pd.concat([class0.iloc[:70], class1.iloc[:140]], ignore_index=True)
test = pd.concat([class0.iloc[70:], class1.iloc[140:]], ignore_index=True)

X_train = train.drop('class', axis=1)
y_train = train['class']
X_test = test.drop('class', axis=1)
y_test = test['class']

print("Training Set Label Distribution:\n", train['class'].value_counts())
print("Training Set Head:\n", test.head())
```

(c)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

k_values = list(range(208, 0, -3))  

train_errors = []
test_errors = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(X_train, y_train)
    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)
    train_errors.append(1 - accuracy_score(y_train, y_train_pred))
    test_errors.append(1 - accuracy_score(y_test, y_test_pred))

plt.figure(figsize=(10, 6))
plt.plot(k_values, train_errors, label='Train Error')
plt.plot(k_values, test_errors, label='Test Error')
plt.xlabel('k')
plt.ylabel('Error Rate')
plt.legend()
plt.title('Train/Test Error vs. k (Euclidean Distance)')
plt.show()
```

```python
best_k = k_values[np.argmin(test_errors)]
print(f"best_k: {best_k}")
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

knn = KNeighborsClassifier(n_neighbors=best_k, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
tpr = tp / (tp + fn)  
tnr = tn / (tn + fp)  
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"confusion_matrix:\n{confusion_matrix(y_test, y_pred)}")
print(f"TPR: {tpr:.4f}, TNR: {tnr:.4f}, Precision: {precision:.4f}, F1: {f1:.4f}")
```

```python
best_test_errors = []

N_values = list(range(10, 211, 10)) 

for N in N_values:

    n_class0 = int(np.floor(N / 3))
    n_class1 = N - n_class0
    

    subset_class0 = class0.iloc[:n_class0]
    subset_class1 = class1.iloc[:n_class1]
    subset_train = pd.concat([subset_class0, subset_class1], ignore_index=True)
    
    X_subset = subset_train.drop('class', axis=1)
    y_subset = subset_train['class']
    

    k_candidates = list(range(1, len(subset_train) + 1, 5))
    errors = []
    
    for k in k_candidates:
        knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        knn.fit(X_subset, y_subset)
        y_pred = knn.predict(X_test)
        errors.append(1 - accuracy_score(y_test, y_pred))
    
    best_test_errors.append(min(errors))


plt.figure(figsize=(10, 6))
plt.plot(N_values, best_test_errors, marker='o')
plt.xlabel('Training Set Size (N)')
plt.ylabel('Best Test Error Rate')
plt.title('Learning Curve (Euclidean Distance)')
plt.show()
```

```python
k_candidates = list(range(1, 197, 5))  
best_k_manhattan = None
best_error_manhattan = 1.0

for k in k_candidates:
    knn = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    knn.fit(X_train, y_train)
    error = 1 - knn.score(X_test, y_test)
    if error < best_error_manhattan:
        best_error_manhattan = error
        best_k_manhattan = k

print(f"best_k_manhattan*: {best_k_manhattan}, best_error_manhattan: {best_error_manhattan:.4f}")
```

```python
import numpy as np

log_p_values = np.arange(0.1, 1.1, 0.1)  
p_values = 10 ** log_p_values  

best_log_p = None
best_error_p = 1.0

for log_p in log_p_values:
    p = 10 ** log_p
    knn = KNeighborsClassifier(
        n_neighbors=best_k_manhattan,  
        metric='minkowski',
        p=p
    )
    knn.fit(X_train, y_train)
    error = 1 - knn.score(X_test, y_test)
    if error < best_error_p:
        best_error_p = error
        best_log_p = log_p

print(f"best_log_p: {best_log_p:.1f}, best_error_p: {best_error_p:.4f}")
```

```python
k_candidates = list(range(1, 197, 5))
best_k_chebyshev = None
best_error_chebyshev = 1.0

for k in k_candidates:
    knn = KNeighborsClassifier(n_neighbors=k, metric='chebyshev')
    knn.fit(X_train, y_train)
    error = 1 - knn.score(X_test, y_test)
    if error < best_error_chebyshev:
        best_error_chebyshev = error
        best_k_chebyshev = k

print(f"best_k_chebyshev*: {best_k_chebyshev}, best_error_chebyshev: {best_error_chebyshev:.4f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

cov_matrix = np.cov(X_train.T)
inv_cov_matrix = np.linalg.pinv(cov_matrix)

best_k_mahalanobis = None
best_error_mahalanobis = 1.0

for k in k_candidates:
    knn = KNeighborsClassifier(
        n_neighbors=k,
        metric='mahalanobis',
        metric_params={'VI': inv_cov_matrix}
    )
    knn.fit(X_train, y_train)
    error = 1 - knn.score(X_test, y_test)
    if error < best_error_mahalanobis:
        best_error_mahalanobis = error
        best_k_mahalanobis = k

print(f"best_k_mahalanobis*: {best_k_mahalanobis}, best_error_mahalanobis: {best_error_mahalanobis:.4f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

metrics = ['euclidean', 'manhattan', 'chebyshev']
best_errors = {}

for metric in metrics:
    best_error = 1.0
    best_k = None

    k_values = list(range(1, 197, 5))
    
    for k in k_values:
        knn = KNeighborsClassifier(
            n_neighbors=k,
            metric=metric,
            weights='distance'  
        )
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        error = 1 - accuracy_score(y_test, y_pred)
        
        if error < best_error:
            best_error = error
            best_k = k
    
    best_errors[metric] = (best_k, best_error)

print("Weighted Voting Best Test Error:")
for metric, (k, error) in best_errors.items():
    print(f"- {metric.capitalize()} metric: k*={k}, test_error={error:.4f}")
```
